```python
import matplotlib.pyplot as plt 
import numpy as np


def plot (x_data,y_data):
        fig = plt.figure() 
        ax = plt.axes() 
        ax.plot(x_data, y_data) 

        ax.grid( linewidth=0.8,
                linestyle='--')
        plt.plot(x_data, y_data, linewidth ='1', color='r')
        plt.show() 


x = np.linspace(0, 20, 1000) 
y = (np.sin(x)/10)+0.9

plot(x,y)

f=1.3
p=3.1413
w=2*p*f

y = (np.sin(x*w)/10)+0.9
plot(x,y)

R = 0.007 * y
plot(x,R)

p1 = 120
p2 = 110
l = 133*0.001
Vyzkost = 0.5

Q = (p*R**4*(p1-p2))/(8*Vyzkost*l)

plot(x,Q)

q=np.mean(Q)
print(q)

```
